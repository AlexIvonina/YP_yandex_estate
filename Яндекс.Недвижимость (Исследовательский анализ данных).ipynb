## Предобработка данных<a id='preprocessing'></a>
###  Заполнение пропусков, проверка на дубликаты

Посмотрим, сколько пропусков есть в таблице и в каких колонках.

```python
# подсчет пропусков и визуализация процентного соотношения пропусков
display(pd.DataFrame(round((df.isna().mean()*100),2).sort_values(ascending=False), columns=['NaNs, %']).style.format(
    '{:.2f}').background_gradient('bwr'))
df.isna().sum().sort_values(ascending=False)
```

Видно, что пропусков много. Удалить мы можем не более 10% из изначального кол-ва 23699.

Больше всего пропусков (88%) в столбце `is_apartment`. Скорее всего пропуски могли возникнуть из-за того, что поле было не обязательным, и пользователи не указывали отрицательные значения. Также нельзя исключить и техническую ошибку в данных.

Следом (66% и 62%) столбцы `parks_nearest` и `ponds_nearest`- Данные об удаленности от парков и водоемов получены из картографии. Чтобы понять, откуда эти  пропуски, посмотрим на населенные пункты с пропущенными значениями и без (все же СПб и ЛО разные субъекты федерации). Также можно предположить, что после определенного расстояния, данные не указываются.

Аналогично с остальными пропусками в картографических данных (по 23% пропусков): `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `ponds_around3000`. Кроме того, процент пропусков почти одинаковый в этих колонках. Надо посмотреть, не одни ли и те же это строки.

Пропуски в данных про балконы `balcony` (49%) - скорее всего это поле при оформлении объявления не было обязательным и, если у пользователя балкона нет, он оставлял это поле пустым, а не ставил "0". Проверим, как наличие балкона зависит от этажа и от расстояния от центра города.

```python
# в df_balcony отфильтруем столбцы
df_balcony=df[['balcony','open_plan','floor','cityCenters_nearest']]
# добавим флаг наличия балкона
df_balcony['balcony_av']=df_balcony['balcony'].isna()==0
# добавим столбец с расстоянием в км
df_balcony['cityCenters_gr']=round(df_balcony['cityCenters_nearest']/1000,0)
# отобразим первые строчки
df_balcony.head()
```

```python
# отобразим график зависимости наличия балкона от этажа
(df_balcony.pivot_table(index='floor', columns='balcony_av', values='open_plan', aggfunc='count')
.plot(kind='bar', figsize=(10,6), title='Наличие балкона по этажам'));
```

```python
# отобразим график зависимости наличия балкона от расстояния до центра
(df_balcony.pivot_table(index='cityCenters_gr', columns='balcony_av', values='open_plan', aggfunc='count')
.plot(kind='line', figsize=(10,6), title='Наличие балкона по удаленности от центра'));
```

По графикам видно, что пропусков в колонке `balcony` больше всего в первых этажах и ближе к центру города. Логично предположить, что балконов там нет и заполнить эти пропуски нулями.

С высотой потолков `ceiling_height` (39%) сложнее. Но природа пропусков, скорее всего именно в необязательности заполнения поля или, как вариант в потере данных. Если поле было не обязательным, то логично предположить, что в момент оформления объявления около 40% не могли быстро посмотреть в документах точную высоту потолков, измерять с линейкой точно никто не хотел/не мог, поэтому предпочли оставить пустое значение. Для удаления пропусков много, а высота потолков может быть параметром для оценки жилья. На практике было бы правильно установить причину пропусков и восстановить данные. Или хотя бы запросить данные о годе постройки дома, его типе планировки для более корректного заполнения. Такой возможности нет в учебном проекте.

Пропуски в технических данных столбца `days_exposition` (13%) - скорее на момент выгрузки эти объявления еще были опубликованы. Посмотрим на даты публикаций этих объявлений.

Пропуски в `kitchen_area` (10%) и `living_area` (8%) - важные показатели для определения стоимости. Возможно это квартиры-студии и владельцы не могли выделить жилую и кухонную площади. Или это ошибки в данных.

Меньше процента пропусков в колонках `floors_total` и `locality_name`. Т.к. процент небольшой, можно удалить эти данные.

```python
# добавляем колонку с процентным соотношением площади кухни к общей.
df['total_kitchen_area_ratio']=df['kitchen_area']/df['total_area']*100
```

Заполним пропуски в этих столбцах (`total_living_area_ratio`,`total_kitchen_area_ratio`) медианой с группировкой по влияющим колонкам `total_area`, `rooms`.

```python
# в total_area_round округляем общую площадь для лучшей группировки
df['total_area_round']=round(df['total_area'],0)
# считаем пропуски в total_living_area_ratio и total_kitchen_area_ratio
df['total_living_area_ratio'].isna().sum(), df['total_kitchen_area_ratio'].isna().sum()
```

```python
# заполняем пропуски total_living_area_ratio медианой по этому столбцу с группировкой по общей площади и кол-ву комнат
df['total_living_area_ratio'].fillna(df.groupby(['total_area_round', 'rooms'])['total_living_area_ratio']\
                                     .transform('median'),inplace = True)
# заполняем пропуски total_kitchen_area_ratio медианой по этому столбцу с группировкой по общей площади и кол-ву комнат
df['total_kitchen_area_ratio'].fillna(df.groupby(['total_area_round', 'rooms'])['total_kitchen_area_ratio']\
                                      .transform('median'),inplace = True)

# считаем пропуски в total_living_area_ratio и total_kitchen_area_ratio
df['total_living_area_ratio'].isna().sum(), df['total_kitchen_area_ratio'].isna().sum()
```

Большую часть пропусков заполнили. Сгуппируем общую площадь с шагом в 20кв.м. и заполним оставшиеся пропуски медианой с группировкой только по общей площади.

```python
# в total_area_round группируем общую площадь с шагом в 20 кв.м.
df['total_area_round']=round(df['total_area']/20,0)
# заполняем пропуски total_living_area_ratio медианой по этому столбцу с группировкой по общей площади
df['total_living_area_ratio'].fillna(df.groupby(['total_area_round'])['total_living_area_ratio']\
                                     .transform('median'),inplace = True)
# заполняем пропуски total_kitchen_area_ratio медианой по этому столбцу с группировкой по общей площади
df['total_kitchen_area_ratio'].fillna(df.groupby(['total_area_round'])['total_kitchen_area_ratio']\
                                      .transform('median'),inplace = True)


# считаем пропуски в total_living_area_ratio и total_kitchen_area_ratio
df['total_living_area_ratio'].isna().sum(), df['total_kitchen_area_ratio'].isna().sum()
```

Пропуски заполнены. Умножая на этот коэффициент общую площадь, заполним пропуски в `living_area` и `kitchen_area`.

```python
# заполняем пропуски в living_area и kitchen_area
df['living_area']=df['total_area']*df['total_living_area_ratio']/100
df['kitchen_area']=df['total_area']*df['total_kitchen_area_ratio']/100
# считаем пропуски в living_area и kitchen_area
df['living_area'].isna().sum(), df['kitchen_area'].isna().sum()
```

```python
# создаем функцию, которая распределяет населенные пункты по типу
SPb_group=(['поселок Шушары', 'поселок Парголово', 'Ломоносов', 'Петергоф',
       'Пушкин', 'Колпино', 'Кронштадт', 'поселок Металлострой',
       'Павловск', 'Сестрорецк', 'Красное Село', 'поселок Понтонный',
       'Зеленогорск', 'поселок Молодежное', 'поселок Стрельна',
       'поселок Александровская', 'поселок Песочный',
       'поселок Петро-Славянка', 'поселок Репино', 'поселок Саперный',
       'поселок Лисий Нос', 'поселок Левашово', 'поселок Белоостров',
       'поселок Усть-Ижора'])

def locality_group(row):
    """
    Функция возвращает тип населенного пункта:
        деревня
        коттеджный поселок
        поселок городского типа
        поселок
        село
        СНТ
        СПб
        город
        в СПб
    
    """
     
    if row in SPb_group:
        return 'в СПб'
    elif 'деревня ' in row:
        return 'деревня'
    elif 'коттеджный поселок' in row:
        return 'коттеджный поселок '
    elif 'поселок городского типа ' in row:
        return 'поселок городского типа'
    elif 'поселок ' in row:
        return 'поселок'
    elif 'село ' in row:
        return 'село'
    elif 'товарищество ' in row:
        return 'СНТ'
    elif 'Санкт-Петербург' in row:
        return 'СПб'
    return 'город'

# создаем столбец locality_group в df
df['locality_group']=df['locality_name'].apply(locality_group)

# смотрим первые 10 строчек df
df.head(10)
```

```python
# создаем функцию для просмотра количественных колонок на предмет аномалий


def describe_hist_boxplot(data, column, range_col=(0,0), bins='auto'):
    """
    Функция возвращает информацию о колонке:
    - описательную статистику
    - гистограмма
    - диаграмму размаха
    
    """

    # построим гистрограмму
    if range_col==(0,0):
        data[column].hist(bins=bins, figsize=(15,8),range=(data[column].min(),data[column].max()))
    else:
        data[column].hist(bins=bins, figsize=(15,8),range=range_col)
    plt.title('Распределение '+column+' \n', fontsize=16)
    plt.show()
    # отобразим описательную статистику
    display(data[column].describe())
    # строим диаграмму размаха
    data.boxplot([column], figsize=(5,10))
    plt.show()
```

```python
# смотрим статистику и графики по колонке rooms
describe_hist_boxplot(df,'rooms', bins=40)
```

## Расчёты и добавление результатов в таблицу

Для определения параметров оценки стоимости квартир, рассчитаем и добавим в таблицу следующие данные:
- цена квадратного метра `square_meter_price`
- день недели, месяц и год публикации объявления `day_start`, `month_start`, `year_start`
- этаж квартиры; варианты — первый, последний, другой `floor_category`

Уже добавлены необходимые столбцы:
- соотношение жилой и общей площади `total_living_area_ratio`
- отношение площади кухни к общей `total_kitchen_area_ratio`

```python
# добавляем колонку square_meter_price в df
df['square_meter_price']=round(df['last_price']/df['total_area'],1)
# создаем столбцы и записываем в них день недели, месяц и год публикации
df['day_start'] = df['first_day_exposition'].dt.dayofweek
df['month_start'] = df['first_day_exposition'].dt.month
df['year_start'] = df['first_day_exposition'].dt.year
```

```python
# создаем функцию, которая распределяет этажность квартиры по типу
def floor_category(row):
    """
    Функция возвращает тип этажа:
        первый
        последний
        другой
    
    """
    floors_total=row['floors_total']
    floor=row['floor']     
    
    if floor==1:
        return 'первый'
    elif floors_total==floor:
        return 'последний'
    return 'другой'

# добавляем колонку floor_category в df
df['floor_category']=df.apply(floor_category,axis=1)

# смотрим первые строчки df
display(df.head())

# смотрим статистику
df['square_meter_price'].describe()
```
